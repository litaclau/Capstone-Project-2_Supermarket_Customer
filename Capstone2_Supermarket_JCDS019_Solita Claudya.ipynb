## 🔄 **DATA PREPARATION**
---

**Berdasarkan hasil tinjauan awal, terdapat beberapa poin penting yang akan ditangani pada tahap ini yakni:**
- Mengkonversi data **Dt_Customer** menjadi datetime dan memisahkannya menjadi tanggal, bulan, tahun
- Menangani Missing Value pada kolom **Income** 
- Menangani outlier pada kolom **Income** dan **Year_Birth**
- Mengkategorikan kolom **Education**, **Marital_Status**, dan **Income** 
- Membuat kolom baru yang bisa menjadi informasi tambahan untuk analisis
- Menghapus kolom yang tidak dibutuhkan dalam proses analisis, yakni kolom **Z_CostContanct**, **Z_Revenue**, **Recency**, **Year_Birth**, 
**Kidhome**, dan **Teenhome** 
 
---

### 1️⃣ Konversi Tipe Data **Dt_Customer**


```python
df["Dt_Customer"] = pd.to_datetime(df["Dt_Customer"], dayfirst=True)
tabel_tipe_data = pd.DataFrame({
    "Kolom": ["Dt_Customer"],
    "Tipe Data": [df["Dt_Customer"].dtype]
})
tabel_tipe_data
```

```python
# Ekstrasi menjadi tanggal, bulan, dan tahun

df["Day_Join"] = df["Dt_Customer"].dt.day
df["Month_Join"] = df["Dt_Customer"].dt.month
df["Year_Join"] = df["Dt_Customer"].dt.year
```

### 2️⃣ Menangani Missing Value & Outlier pada **Income**

```python
median_Pendapatan = df["Income"].median()
print(median_Pendapatan)
df.fillna({"Income": median_Pendapatan}, inplace=True)

missing_pendapatan = pd.DataFrame({
    "Kolom": ["Income"],
    "Jumlah Missing": [df["Income"].isnull().sum()],
})
missing_pendapatan
```

```python
# Menggunakan Interquartile Range (IQR)
Q1 = df['Income'].quantile(0.25)
Q3 = df['Income'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df = df[(df['Income'] >= lower_bound) & (df['Income'] <= upper_bound)]

print("Data sesudah tanpa outlier :", df.shape)

df.describe().transpose()
```

### 3️⃣ Menangani Outlier pada **Year_Birth**

Berbeda halnya dengan outlier pada Income, outlier pada Year_Birth tidak dihapus melainkan mengganti outlier tersebut dengan nilai median dari Year_Birth. Hal ini dilakukan guna mengurangi dampak outlier dan membuat distribusi data lebih representatif untuk analisis selanjutnya.

```python
jumlah_bawah_1940 = df[df['Year_Birth'] < 1940].shape[0]
jumlah_atas_1940 = df[df['Year_Birth'] >= 1940].shape[0]

info_df = pd.DataFrame({
    'Tahun Lahir': ['< 1940', '≥ 1940'],
    'Jumlah': [jumlah_bawah_1940, jumlah_atas_1940]
})

info_df
```

```python
# Menghitung Nilai Median dari Tahun_Lahir >= 1940
median_year = df[df['Year_Birth'] >= 1940]['Year_Birth'].median()
print(f"Median Tahun Lahir (≥ 1940): {median_year}")

# Mengubah 3 Outlier menjadi nilai Median
df.loc[df['Year_Birth'] < 1940, 'Year_Birth'] = median_year

# Visualisasi
plt.figure(figsize=(10, 4))
sns.histplot(df['Year_Birth'], bins=30, kde=True)
plt.title('Distribusi Tahun Lahir (Setelah Penanganan Outlier)')
plt.xlabel('Tahun Lahir')
plt.ylabel('Jumlah')
plt.show()
```

### 4️⃣ Mengkategorikan ulang isi kolom **Education**, **Marital_Status**, dan **Income** 

```python
# Pengkategorian Education

df["Education"] = df["Education"].replace({
    "Basic"     : "Basic",
    "Graduation": "Undergraduate",
    "2n Cycle"  : "Postgraduate",
    "Master"    : "Postgraduate",
    "PhD"       : "Postgraduate"
})

tabel_pendidikan = df['Education'].value_counts().reset_index()
tabel_pendidikan.columns = ['Education', 'Jumlah']
tabel_pendidikan
```

```python
# Pengkategorian Marital_Status

df['Marital_Status'] = df['Marital_Status'].replace({
    'Married'   : 'Married',
    'Together'  : 'Married',
    'Single'    : 'Unmarried',
    'Divorced'  : 'Unmarried',
    'Widow'     : 'Unmarried',
    'Alone'     : 'Unmarried',
    'YOLO'      : 'Unmarried',
    'Absurd'    : 'Unmarried'
})
status_tabel = df['Marital_Status'].value_counts().reset_index()
status_tabel.columns = ['Marital_Status', 'Jumlah']
status_tabel
```

```python
# Pengkategorian Income

def kategori_pendapatan(x):
    if x < 40000:
        return 'Low'
    elif x <= 80000:
        return 'Middle'
    else:
        return 'High'

df['Income Category'] = df['Income'].apply(kategori_pendapatan)

Income_Category = df['Income Category'].value_counts().reset_index()
Income_Category.columns = ['Income Category', 'Jumlah']
Income_Category
```

### 5️⃣ Membuat Kolom Baru yang bisa menjadi Informasi Tambahan untuk Analisis

#### **Generation**

```python
def get_generasi(tahun):
    if 1946 <= tahun <= 1964:
        return 'Baby Boomer'
    elif 1965 <= tahun <= 1980:
        return 'Gen X'
    elif 1981 <= tahun <= 1996:
        return 'Millennial'
    else:
        return 'Pre Baby Boomer'

df['Generation'] = df['Year_Birth'].apply(get_generasi)

df['Generation'].value_counts()
```

#### **Family_Size**

```python
df['Family_Size'] = df['Marital_Status'].replace({
    'Married': 2,
    'Unmarried': 1
}) + df['Kidhome'] + df['Teenhome']

df[['Family_Size']].head()
```

#### **T_AmountSpent**

```python
produk = ['Wine', 'Fruit', 'Meat', 'Fish', 'Sweet', 'Gold']

df['T_AmountSpent'] = df[produk].sum(axis=1)
df[['T_AmountSpent']].head()
```

#### **T_Purchase**

```python
Sarana_Pembelian = ['Web_Purchase', 'Catalog_Purchase', 'Store_Purchase']

df['T_Purchase'] = df[Sarana_Pembelian].sum(axis=1)
df[['T_Purchase']].head()
```

#### **T_AcceptedCampaign**

```python
Campaign = [
    'Campaign1', 'Campaign2', 'Campaign3',
    'Campaign4', 'Campaign5', 'Latest_Campaign'
]

df['T_AcceptedCampaign'] = df[Campaign].sum(axis=1)
df[['T_AcceptedCampaign']].head()
```

### 7️⃣ Menghapus kolom yang tidak dibutuhkan dalam proses analisis, yakni kolom **Z_CostContanct**, **Z_Revenue**, **Year_Birth**, **KidHome**, **TeenHome**

```python
df.drop(columns=["Z_CostContact", "Z_Revenue","Recency", "Year_Birth", "Kidhome", "Teenhome"], inplace=True)
```

> 5. Apakah ada korelasi antara pendapatan (Income) dengan jumlah pengeluaran pelanggan (T_AmountSpent)?

```python
from scipy.stats import pearsonr

# Asumsi: dataframe kamu bernama df dan sudah berisi kolom Income (nominal) dan T_AmountSpent

# 1. Hitung korelasi Pearson antara Income dan T_AmountSpent
corr, p_value = pearsonr(df['Income'], df['T_AmountSpent'])

print(f"Korelasi Pearson antara Income dan T_AmountSpent: r = {corr:.3f}, p-value = {p_value:.3e}")

# 2. Scatter plot Income vs T_AmountSpent dengan nilai korelasi di plot
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='Income', y='T_AmountSpent', alpha=0.6)


plt.text(x=df['Income'].min(), y=df['T_AmountSpent'].max()*0.9,
         s=f'r = {corr:.3f}\np = {p_value:.3e}',
         fontsize=12, bbox=dict(facecolor='white', alpha=0.5))

plt.title('Scatter Plot Income vs Total Amount Spent')
plt.xlabel('Income (Nominal)')
plt.ylabel('Total Amount Spent')
plt.tight_layout()
plt.show()

# 3. Boxplot T_AmountSpent per Income Category dengan annotasi nilai median di atas boxplot
plt.figure(figsize=(8,6))
ax = sns.boxplot(data=df, x='Income Category', y='T_AmountSpent', order=['Low', 'Middle', 'High'])


medians = df.groupby('Income Category')['T_AmountSpent'].median().reindex(['Low', 'Middle', 'High'])
positions = range(len(medians))
for pos, median in zip(positions, medians):
    ax.text(pos, median + df['T_AmountSpent'].max()*0.02, f'{median:.0f}', 
            ha='center', va='bottom', fontsize=11, color='black')

plt.title('Boxplot Total Amount Spent per Income Category')
plt.xlabel('Income Category')
plt.ylabel('Total Amount Spent')
plt.tight_layout()
plt.show()
```

#### **Berdasarkan Visualisasi Scatter Plot**
- Koefisien Korelasi Pearson antara Income dan Total Amount Spent adalah **r = 0.820** dengan p-value < 0.001
- Terdapat **hubungan positif yang sangat kuat dan signifikan secara statistik** antara pendapatan pelanggan (Income) dengan jumlah pembelian (Total Amount Spent)
- Hal ini juga divisualisasikan oleh pola Scatter plot yang terus naik seiring dengan peningkatkan Income

#### **Berdasarkan Visualisasi Boxplot**
- Menunjukkan peningkatan nilai median seiring dengan peningkatan kategori Income pelanggan 
- Income Low    : Distribusi sangat sempit artinya sebagian besar pelanggan dengan income low tidak banyak membeli produk
- Income Middle : Distribusi lebih lebar artinya sebagian besar pelanggan dengan income middle lebih banyak membeli produk namun terdapat beberapa outlier sehingga tidak sestabil income high
- Income High   : Distribusi pembelian lebar dan stabil (minim outlier) artinya pelanggan dengan income high konsisten membeli produk dengan jumlah yang tinggi  

#### **Kesimpulan**
- Terdapat perbedaan signifikan antar kelompok Income
- Semakin tinggi pendapatan pelanggan maka semakin besar jumlah pembelian produk 
- Pelanggan dengan Income High cenderung lebih konsisten membeli produk dengan jumlah yang tinggi

```python
# Membuat salinan DataFrame platform pembelian + kategorisasi
df_platform = df[['Generation', 'Income Category', 'Marital_Status', 'Family_Size', 'Education',
                  'Web_Purchase', 'Catalog_Purchase', 'Store_Purchase']]

# ------------------ AGREGASI ------------------ #
purchase_by_generation = df_platform.groupby('Generation')[['Web_Purchase', 'Catalog_Purchase', 'Store_Purchase']].sum()
purchase_by_income = df_platform.groupby('Income Category')[['Web_Purchase', 'Catalog_Purchase', 'Store_Purchase']].sum()
purchase_by_marital = df_platform.groupby('Marital_Status')[['Web_Purchase', 'Catalog_Purchase', 'Store_Purchase']].sum()
purchase_by_family = df_platform.groupby('Family_Size')[['Web_Purchase', 'Catalog_Purchase', 'Store_Purchase']].sum()
purchase_by_education = df_platform.groupby('Education')[['Web_Purchase', 'Catalog_Purchase', 'Store_Purchase']].sum()

# ------------------ FUNGSI UNTUK PLOTTING ------------------ #
def plot_grouped_bar(data, title, xlabel, figsize=(10,6), colormap='viridis'):
    ax = data.plot(kind='bar', stacked=False, figsize=figsize, colormap=colormap)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel('Jumlah Pembelian')
    plt.xticks(rotation=0)
    plt.legend(title='Platform Pembelian')
    for container in ax.containers:
        ax.bar_label(container, label_type='edge', fontsize=9, padding=3)
    plt.tight_layout()
    plt.show()

# ------------------ PLOT SEMUA ------------------ #
plot_grouped_bar(purchase_by_generation, 
                 'Total Pembelian per Platform Berdasarkan Generation', 
                 'Generation')

plot_grouped_bar(purchase_by_income, 
                 'Total Pembelian per Platform Berdasarkan Income Category', 
                 'Income Category', 
                 colormap='viridis')

plot_grouped_bar(purchase_by_marital, 
                 'Total Pembelian per Platform Berdasarkan Marital Status', 
                 'Marital Status', 
                 colormap='viridis')

plot_grouped_bar(purchase_by_family, 
                 'Total Pembelian per Platform Berdasarkan Family Size', 
                 'Family Size', 
                 colormap='viridis')

plot_grouped_bar(purchase_by_education, 
                 'Total Pembelian per Platform Berdasarkan Education', 
                 'Education', 
                 colormap='viridis')
```

> 12.  Segmen pasar mana yang paling potensial untuk ditargetkan di masa depan?

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 1. Membentuk kolom Segment gabungan dari Generation dan Income Category
df['Segment'] = df['Generation'] + ' - ' + df['Income Category']

# 2. Menghitung skor potensi per segmen (misalnya total pembelian atau skor lainnya)
segment_summary = df.groupby('Segment').agg({
    'T_AmountSpent': 'sum'  # atau ganti dengan kolom lain sebagai Potential_Score
}).reset_index().rename(columns={'T_AmountSpent': 'Potential_Score'})

# 3. Sort berdasarkan skor tertinggi
segment_summary_sorted = segment_summary.sort_values(by='Potential_Score', ascending=False)

# 4. Data untuk pie chart
sizes = segment_summary_sorted['Potential_Score']
labels = segment_summary_sorted['Segment']

# 5. Pie chart dengan anotasi elegan
fig, ax = plt.subplots(figsize=(10,10))
wedges, texts, autotexts = ax.pie(
    sizes,
    labels=None,
    autopct='%1.1f%%',
    startangle=140,
    pctdistance=0.75,
    textprops={'fontsize': 10}
)

for i, p in enumerate(wedges):
    ang = (p.theta2 + p.theta1) / 2.
    x = np.cos(np.deg2rad(ang))
    y = np.sin(np.deg2rad(ang))

    label_x = 1.15 * x
    label_y = 1.15 * y

    horizontal_alignment = 'left' if x > 0 else 'right'
    connectionstyle = f"angle,angleA=0,angleB={ang}"

    ax.annotate(
        labels.iloc[i],
        xy=(x * 1.0, y * 1.0),
        xytext=(label_x, label_y),
        ha=horizontal_alignment,
        va='center',
        arrowprops=dict(arrowstyle='-', connectionstyle=connectionstyle, color='black', lw=1),
        fontsize=10,
        bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="gray", lw=0.5)
    )

plt.title('Distribusi Potensi Segmen Pasar Berdasarkan Generasi dan Income', fontsize=14)
plt.axis('equal')
plt.tight_layout()
plt.show()
```

Diagram distribusi total Pembelian per Segmen berdasarkan Generasi dan Income menunjukan bahwa **lebih dari 60% total pembelian berasal dari segmen Gen X - Middle (31.8%) dan Baby Boomer - Middle (28.4%)**. Hal ini berarti pelanggan yang generasinya lebih tua dengan pendapatan menengah adalah kontributor utama dalam penjualan.

## 🎯 **Rekomendasi**

---

#### **Meningkatkan Penjualan Produk melalui Campaign dengan Sasaran dan Strategi Campaign yang tepat**
##### **`Menentukan Target Pasar/Sasaran yang tepat berdasarkan Tujuan Campaign`**

**Gen X, Baby Boomer, Middle Income, dan Married** merupakan kelompok pelanggan yang tepat untuk menjadi target campaign peningkatan penjualan. Secara khusus, kelompok dengan segmentasi **Gen X - Middle** dan **Baby Boomer - Middle**. Selama 2 tahun terakhir kelompok ini menunjukkan sikap yang loyal (jumlah pembelian tinggi dan stabil) terhadap LTS Market. Tingginya jumlah produk yang dibeli oleh kelompok ini menunjukan bahwa kelompok ini adalah **target yang tepat untuk Campaign Upselling**.

Beberapa penawaran yang bisa ditawarkan pada kelompok ini adalah member **rewards, referral bonus, Campaign eksklusif, early access, atau VIP treatment**. Tim pemasaran bisa menetapkan jumlah minimal belanja untuk bisa mendapatkan manfaat eksklusif tersebut. 
Sebagai contoh, member eksklusif bisa mendapat early access terhadap produk baru atau jaminan ketersediaan stok produk best seller seperti Wine dan Meat. 

Selain itu, kelompok dengan segmentasi **Baby Boomer - High**, **Millennial - Middle**, **Gen X - High** juga dapat dijadikan target untuk Campaign Upselling. Walaupun jumlah pembelian produknya belum setinggi pelanggan yang loyal, namun kelompok ini dapat meningkatkan penjualan dalam jangka panjang.

##### **`Membuat Strategi Campaign yang Efektif serta sesuai dengan Tujuan dan Sasaran Campaign`**
- **Mengadaptasi dan mengembangkan strategi campaign terakhir** untuk campaign masa mendatang. Oleh karena hasil analisis menunjukkan bahwa terdapat peningkatan efektivitas atau frekuensi Campaign pada campaign terakhir. 
- Membuat strategi campaign yang **cocok dengan karakteristik pelanggan**. Dalam case ini, LTS Market perlu membuat campaign yang mudah diterima oleh Generasi X, dapat dijangkau oleh pelanggan dengan penghasilan menengah, dapat diterima oleh kelompok terdidik, cocok untuk keluarga.
- Memberikan penawaran-penawaran khusus untuk pembelian pertama agar meningkatkan minat kelompok yang belum pernah menerima campaign  

##### **`Menambah Data Pendukung untuk Analisis Keberhasilan Campaign`**
Tren pendaftaran member baru dapat dianalisis berdasarkan data pada kolom **Dt_Customer**, namun data ini akan lebih informatif lagi jika dibandingkan dengan waktu pelaksanaan campaign. Data ini dapat menjadi evaluasi keberhasilan dari campaign yang dilakukan oleh tim pemasaran. Selain itu, tim pemasaran perlu memperkuat promosi pada bulan-bulan di awal tahun agar tren member tetap stabil dan justru meningkat. 

#### **Membuat Strategi untuk Meningkatkan Penjualan untuk Produk Gold, Fish, Sweet, dan Fruit**
- Melakukan **Cross-selling** dapat menjadi salah satu solusi agar penjualan produk lebih merata dan tidak terfokus hanya di produk Wine dan Meat 
- Menawarkan paket bundle produk Gold, Fish, Sweet, dan Fruit dengan produk yang tingkat penjualannya tinggi seperti Wine dan Meat
- **Paket bundle dibuat bergilir** dari satu kategori produk ke kategori produk lain dalam jangka waktu tertentu. Sebagai contoh, minggu pertama paket bundle Wine dan Sweet dan seterusnya
- Solusi ini dapat meningkatkan perputaran stok dan mencegah pembusukan produk segar seperti Fish dan Fruit yang mana masa simpannya pendek

#### **Mengembangkan Platform Penjualan Online**

Pelanggan yang hanya terfokus pada salah satu platform belanja (toko offline), membawa dampak yang kurang baik untuk Market. Investasi/biaya yang dikeluarkan untuk membangun dan memelihara website dan katalog tidak sebanding dengan hasilnya. Selain itu, market akan kesulitan untuk mengukur minat perilaku dan segmentasi pelanggan karena data pelanggan sulit didapatkan. Dampak lain yang berkaitan dengan penjualan adalah jangkauan penjualan yang terbatas wilayah sehingga sulit untuk mendapat pelanggan baru.

Dalam kasus ini, LTS market didominasi oleh **Gen X dan Baby Boomer yang cenderung kurang memahami digital platform** seperti web dan katalog. Oleh karena itu, cara yang mungkin bisa dilakukan yaitu melakukan promosi web dan katalog melalui petugas yang ada di market, misal penjaga kasir. Penjaga kasir dapat menginfokan bahwa terdapat diskon dan loyalty points khusus untuk pembelian melalui website/katalog.

**Adapun beberapa cara lain yang bisa dilakukan, yakni:**
- meningkatkan jumlah stok produk Best Seller di website dan katalog dibandingkan di toko. Melalui cara ini, pelanggan yang belum pernah mencoba    
- mengadakan peluncuran produk baru yang hanya bisa dibeli di website/katalog
- mengunakan push notification melalui website atau email pelanggan untuk mendorong traffic digital dan minat untuk berbelanja
- memperkuat promosi katalog dan website di toko offline baik melalui media cetak maupun media elektronik di toko
- menyediakan fitur khusus untuk integrasi antar platform belanja, dalam hal ini produk yang dibeli melalui katalog dan website dapat diambil di toko
